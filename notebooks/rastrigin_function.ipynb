```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
import matplotlib


matplotlib.rcParams['xtick.direction'] = 'out'
matplotlib.rcParams['ytick.direction'] = 'out'
```

```python
def rastrigin(amplitude=10, n=2, resolution=100, a_min=-5.12, a_max=5.12):
    xs = np.array([np.linspace(a_min, a_max, resolution)  for _ in range(n)])
    xs = np.array(np.meshgrid(*xs))
    ys = amplitude * n + np.sum(xs ** 2 - amplitude * np.cos(2 * np.pi * xs), axis=0)
    return xs, ys
```

```python
fig = plt.figure(figsize=(16, 16))
ax = fig.gca()
# ax = fig.gca(projection='3d')

kek = np.linspace(-5, 5, 10)
kek2 = np.linspace(-5, 5, 10)
wolol = np.stack(np.meshgrid(kek, kek2))
print(wolol.shape)

(x1, x2), y = rastrigin(n=2, resolution=1000)
surf = ax.contour(x1, x2, y, cmap=cm.coolwarm, levels=range(0, 80, 4))
surf = ax.scatter(wolol[0], wolol[1], marker='x', linewidths=2, color='red', s=100)
rand_x, rand_y = np.random.uniform(-1, 1, size=(2, 10, 10))
print(wolol[1].shape)
print(wolol[0].shape)
# ax.arrow(wolol[1], wolol[0], rand_x, rand_y, head_width=0.05, head_length=0.1, fc='k', ec='k')
for y in range(10):
    for x in range(10):
        pos = wolol[:, y, x]
        rand_x, rand_y = np.random.uniform(-1, 1, size=(2,))
        
        ax.arrow(pos[1], pos[0], rand_x, rand_y, head_width=0.05, head_length=0.1, fc='k', ec='k')
# ax.set_zlim(0.01, 100.01)
# ax.zaxis.set_major_locator(LinearLocator(10))
# ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
# fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
print(Z.shape)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
